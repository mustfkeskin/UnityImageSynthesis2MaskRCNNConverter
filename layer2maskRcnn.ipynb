```python
import cv2
import numpy as np
import pandas as pd

layer = cv2.imread("SampleScene_11_layer.png")
seg = cv2.imread("SampleScene_11_id.png")
img = cv2.imread("SampleScene_11_img.png")
```

```python
def get_uniques(image, df):
    
    r = image[:,:,0].ravel()
    g = image[:,:,1].ravel()
    b = image[:,:,2].ravel()

    df["r"] = r
    df["g"] = g
    df["b"] = b

    df = df.groupby(["r","g","b"]).size().reset_index()[["r","g", "b"]]
    return df

def get(image, triple):
    r, g, b = triple
    red = image[:,:,0] == r 
    green = image[:,:,1] == g
    blue = image[:,:,2] == b

    red = red.ravel()
    green = green.ravel()
    blue =  blue.ravel()
    
    dff = pd.DataFrame()
    
    dff["red"] = red
    dff["green"] = green
    dff["blue"] = blue
    
    indexes = np.logical_and(np.logical_and(red, green), blue)
    indexes = indexes.astype(np.uint8) 
    indexes = np.reshape(indexes, (image.shape[0], image.shape[1]))
    return indexes

df = pd.DataFrame()
```

```python
layer_unique = get_uniques(layer, df)
seg_unique = get_uniques(seg, df)
```

```python
for i in range(seg_unique.shape[0]):
    index = get(image=seg, triple=seg_unique.loc[i,:].values)
    cv2.imwrite(str(i)+".jpg", index * 255)
    #cv2.imshow("asd",index * 255)
    #cv2.waitKey(0)
    
#cv2.destroyAllWindows()
```
